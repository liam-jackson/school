```python
# excise the columns of interest to lists  - following what Geron does in Ch4 (pdf page 159) edition 1
X = df['radius_mean'].values    # X  holds radius values
y = (pd.factorize(df.diagnosis)[0]-1)*(-1)   # malignant is 1, benign is 0

X = X.reshape(-1,1)
y = y.reshape(-1,1)

######################### Train a Logistic Regression model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_reg = LogisticRegression()
model = log_reg.fit(X,y)

X_new = np.linspace(min(X), max(X), 5000)   # create an array with possible radius values
count = 0;     # use count to index into X_new and test values to find the 50% threshold
for ii in range(X_new.shape[0]-1):
    a = X_new[count];
    b = X_new[count+1]
    mat = log_reg.predict([a,b])
    count = count+1;
    if mat[0] == 0 and mat[1] == 1:  # save the index at which the classification changes due to passing 50% probability
        threshold_ind = count-1;  # -1 bc we already did count+1 above
threshold = X_new[threshold_ind];
print('The threshold value for the 50% boundary is: ', threshold)

#for ii in range(569):
#    jj = X[ii+1];
#    storage = log_reg.predict([X[ii],X[jj]])
```

```python
bound = np.arange(5, 30.5, 0.5) #array of boundary points to test
sensi = speci = np.reshape(np.zeros(len(bound)), (len(bound),1)) #preallocate

for ii in range(len(bound)):
    decision = bound[ii]  # take the boundary value at index ii and use it as the decision boundary
    wrongly_pos = rightly_pos = wrongly_neg = rightly_neg = allpos = allneg = 0;  #preallocate
    for jj in range(len(y)): # use length of data &diagnosis from dataset
        if X[jj] >= decision and y[jj] == 1:
            rightly_pos= rightly_pos+1;
        elif X[jj] >= decision and y[jj] == 0:
            wrongly_pos = wrongly_pos+1;
        elif X[jj] <= decision and y[jj] == 1:
            wrongly_neg = wrongly_neg+1;
        elif X[jj] <= decision and y[jj] == 0:
            rightly_neg = rightly_neg+1;
    print(ii, decision, wrongly_pos , rightly_pos , wrongly_neg , rightly_neg)    # check your iterations
    allpos = rightly_pos + wrongly_neg
    allneg = rightly_neg + wrongly_pos
    sensi[ii] = rightly_pos/allpos
    speci[ii] = rightly_neg/allneg

# Plot

plt.figure(figsize=(10,5)) # make figure larger
plt.scatter(1-speci,sensi)   # NOTE: 1-TrueNegRate = 1-specificity = FalsePosRate
plt.title('ROC curve',fontsize=22)
plt.xlabel('False Positive Rate',fontsize=18)
plt.ylabel('True Positive Rate',fontsize=18)
plt.yticks(fontsize = 15)
plt.xticks(fontsize = 15)
plt.show()
#print(rightly_pos , wrongly_pos , rightly_neg , wrongly_neg)
```
